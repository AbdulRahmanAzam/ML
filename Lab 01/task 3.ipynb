```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# // LAB 1 QUESTION 3
df = pd.read_csv("cancer patient data sets.csv")

# 📌 A. Perform Graphical Visualization
sns.pairplot(df)
plt.show()
```

```python
# Heat map of missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# B: Display a Correlation Graph
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
# C. Display Distribution of Age and Frequent Cold
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')

plt.subplot(1,2,2)
sns.countplot(data=df, x="Frequent Cold", palette='Set2')

plt.show()
```

```python
# D. Find Outliers in Obesity Feature using Graphical Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Obesity'], color='lightblue')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x=df['Obesity'], color='lightgreen')
plt.show()
```
