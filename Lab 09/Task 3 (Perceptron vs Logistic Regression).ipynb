```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import os
# Download latest version
path = kagglehub.dataset_download("himanshunakrani/iris-dataset")

print("Path to dataset files:", path)
print(os.listdir(path))

df = pd.read_csv(r"C:\Users\azama\.cache\kagglehub\datasets\himanshunakrani\iris-dataset\versions\1\iris.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.metrics import accuracy_score

X = df.drop('species', axis=1)
y = df['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.transform(Xtest)

# Logistic Regression 
logr = LogisticRegression(solver='lbfgs', max_iter=5000, multi_class='auto')
logr.fit(Xtrain, ytrain)

logrtrain_acc = accuracy_score(ytrain, logr.predict(Xtrain))
logrtest_acc = accuracy_score(ytest, logr.predict(Xtest))

# PERCEPTRON
percep = Perceptron(max_iter=5000, tol=1e-3, random_state=42)
percep.fit(Xtrain, ytrain)
perceptrain_acc = accuracy_score(ytrain, percep.predict(Xtrain))
perceptest_acc = accuracy_score(ytest, percep.predict(Xtest))

# Comparing Results
print(f"Model: Logistic Regression, Train Acc: {logrtrain_acc}, Test Acc: {logrtest_acc}")
print(f"Model: Perceptron,          Train Acc: {perceptrain_acc}, Test Acc: {perceptest_acc}")
```

```python
# the difference between perceptron and logistic regresssion

# percentron is simple and only cares about misclassified points, doesnt produce probabilites
# linear classifier, hard class label, updates weights on misclassified samples

# logistic regression predicts probabilites, can be regularized, so performs better on small/medium datasets
# linear classifier with probabilistic output, logistic loss functoin, works even if class are not perfectly seperable
```
