```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv(r"weather_forecast.csv")
df2 = pd.read_csv("solar_observed.csv")

display(df1.head())
display(df2.head())

print(df1.isnull().sum())
print('\n')
print(df2.isnull().sum())
```

```python
df1.drop(columns = ['sky_cover'], inplace=True)
```

```python
for col in ['precip_prob', 'temperature_C', 'dew_point_C', 'relative_humidity', 'wind_speed_m_s']:
    df1[col] = df1[col].fillna(df1[col].mean())
df1.isnull().sum()
```

```python
df2['Solar'] = df2['Solar'].fillna(df2['Solar'].mean())

df2.drop('timestamp', axis=1, inplace=True)
print(df2.isna().sum())
```

```python
# renamiong the column name of df2 from local time to timestamp

df2 = df2.rename(columns={'Local_Time':'timestamp'})
df2.head()
```

```python
df1['timestamp'] = pd.to_datetime(df1['timestamp'], errors='coerce', utc=True)
df2['timestamp'] = pd.to_datetime(df2['timestamp'], errors='coerce', utc=True)

display(df1.head(10))
display(df2.head(10))
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
df1 = df1.dropna(subset=['timestamp'])
df2 = df2.dropna(subset=['timestamp'])
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
df = pd.merge(df1, df2, on='timestamp', how='inner')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['hour'] = df['timestamp'].dt.hour
df['dayofyear'] = df['timestamp'].dt.dayofyear
```

```python
X = df.drop(columns=['timestamp', 'Solar'])
y = df['Solar']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test, timestamp_train, timestamp_test = train_test_split(X, y, df['timestamp'].values, test_size=0.2, random_state=42, shuffle=False)
```

```python
from sklearn.preprocessing import StandardScaler

# numeric_cols = X.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
X = scaler.fit_transform(X)
# X_test = ss.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_yPred = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_yPred = rf.predict(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
# scaling all the features except target variable
# numeric_cols = X.select_dtypes(include=[np.number]).columns
# ss = StandardScaler()
# X_train = ss.fit_transform(X_train)
# X_test = ss.transform(X_test)


svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
svr_yPred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def eval_model(model, y_true, y_pred):
    # Evaluation metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"Model: {model} - MAE: {round(mae,2)}, RMSE:{round(rmse,2)}, R2 Score: {round(r2, 2)}")
    # return mae, rmse
```

```python
eval_model("Linear Regression", y_test, lr_yPred)
eval_model("SVR              ", y_test, svr_yPred)
eval_model("Random Forest    ", y_test, rf_yPred)
```
